```python
import geopandas as gpd
import pandas as pd
import os
```

```python
# Replace 'path/to/your/shapefile.shp' with the actual path to your shapefile
shapefile_path = 'C:/Users/jespi/liberty_station_update/weather-station-optimization/weather_stations/liberty_24_stations'
gdf = gpd.read_file(shapefile_path)

# Print the first few rows of the GeoDataFrame
gdf.head()
```

```python
gdf.columns
```

```python
# Rename the relevant columns
gdf.rename(columns={
    'Station_Na': 'Station.Name',
    'Latitude': 'lat',
    'Longitude': 'lon'
}, inplace=True)

# Add new columns with NA values
new_columns = {
    'Mesonet': pd.NA, 
    'Station.ID': pd.NA,
    'Elevation': pd.NA,
    'County': pd.NA, 
    'First.Data.Date': pd.NA, 
    'Last.Data.Date': pd.NA,
    'Status': pd.NA, 
    'last.year': pd.NA, 
    'Quality': 2, 
    'Tier': 'IOU'
}

for col, value in new_columns.items():
    gdf[col] = value

# Drop the 'asset_type' column if it exists
if 'asset_type' in gdf.columns:
    gdf.drop(columns=['asset_type'], inplace=True)
```

```python
# Define the directory and file name
output_dir = 'C:/Users/jespi/liberty_station_update/weather-station-optimization/weather_stations'
output_file = 'liberty_processed_stations_2024.csv'
output_path = os.path.join(output_dir, output_file)

# Ensure the directory exists
os.makedirs(output_dir, exist_ok=True)

# Save the DataFrame to a CSV file
gdf.to_csv(output_path, index=False)

print(f"DataFrame saved to {output_path}")
```
